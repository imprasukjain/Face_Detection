```python
import numpy as np
import cv2
```

```python
haar = cv2.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
```

```python
#For Real time detection change path of video with 0

cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    if ret == False:
        break
    image = frame.copy()
    
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    
    face_detect = haar.detectMultiScale2(gray,2,5)
    
    for x,y,w,h in face_detect[0]:
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow('frame',image)
    num_faces = len(face_detect[0])
    print(f"Number of faces detected: {num_faces}")
    
    if cv2.waitKey(1) == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
